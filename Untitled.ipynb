```python
import pandas as pd
import numpy as np


df = pd.read_excel (r'Master.xlsx')
df.tail()
```

```python
df2 = pd.read_excel (r'Blocks.xlsx')
df2.columns = ["Block", "Day", "Start", "End"]
df2["Start"]=df2["Start"].apply(str)

for time in df2['Start']:
    if  str(time).endswith(":00"): 
        if int(str(time)[:2]) <12:
            df2['Start'] = df2['Start'].replace([time],str(time)[:-3] + " AM")
        if int(str(time)[:2]) >=12:
            if int(str(time)[:2]) ==12:
                
                df2['Start'] = df2['Start'].replace([time],str(time)[:-3] + " PM")
            else:
                
                hours = int(str(time)[:2]) -12
                hours = str(hours) + str(time)[2:]
                df2['Start'] = df2['Start'].replace([time],hours[:-3] + " PM")


for time in df2['End']:
    if  str(time).endswith(":00"): 
        if int(str(time)[:2]) <12:
            df2['End'] = df2['End'].replace([time],str(time)[:-3] + " AM")
            
        if int(str(time)[:2]) >=12:
            if int(str(time)[:2]) ==12:
                df2['End'] = df2['End'].replace([time],str(time)[:-3] + " PM")
            else:
                hours = int(str(time)[:2]) -12
                hours = str(hours) + str(time)[2:]
                df2['End'] = df2['End'].replace([time],hours[:-3] + " PM")

df2
```

## What is the most busiest rooms in Main building?

```python
df['Loc'] = df['Loc'].str.strip()
df['Rm'] = df['Rm'].str.strip()

busiest = df[df['Bldg'] == "MAIN"]
busiest = busiest[['Loc', 'Rm']].agg(lambda x: ','.join(x.values), axis=1).T.value_counts().head(10)
busiest
```

## What time there are classes in B,107?

## Are any Classes Double Booked?

```python
# Title is diferent but Day Start End Loc Rm is the same
double_booked = df[df['Bldg'] == "MAIN"]

double_booked = double_booked[double_booked.duplicated(['Day', 'Start','Loc','Rm'],keep=False)]
double_booked = double_booked[['Title','Instructor','Day','Start','End','Loc','Rm']] 
double_booked = double_booked.drop_duplicates(keep=False, inplace=False)
double_booked.sort_values(by=['Start']) 
```
