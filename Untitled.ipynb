## Are any Classes Conflicting?

```python
# Bldg Day Start End Loc Rm is conflicting
all_conflicts = pd.DataFrame()

classes = df[df['Block'] != 'Z']
classes = classes.drop_duplicates(['Title','Day','Start','End','Bldg','Loc','Rm'])

for i, row in classes.iterrows():
    conflicts = classes[(classes['Bldg'] == row['Bldg']) & \
                        (classes['Day'] == row['Day']) & \
                        (classes['Start'] <= row['End']) & \
                        (classes['End'] >= row['Start']) & \
                        (classes['Loc'] == row['Loc']) & \
                        (classes['Rm'] == row['Rm']) \
                            ]
    if conflicts.shape[0] > 1:
        all_conflicts = all_conflicts.append(conflicts, ignore_index=True)

all_conflicts = all_conflicts.drop_duplicates()
all_conflicts.sort_values(by=['Start', 'Loc', 'Rm']) 
```

## How many hours is Bldg Main, Loc F, Rm 111 used per week?

```python
room = df[(df['Bldg'] == "MAIN") & (df['Loc'] == "F") & (df['Rm'] == "111")]

seconds = 0
for i, row in room.iterrows():
    s = datetime.datetime.combine(datetime.date.min, row.Start)
    e = datetime.datetime.combine(datetime.date.min, row.End)
    seconds += pd.Timedelta(e - s).seconds

hours = seconds / 3600
print(hours, "hours per week")
```
